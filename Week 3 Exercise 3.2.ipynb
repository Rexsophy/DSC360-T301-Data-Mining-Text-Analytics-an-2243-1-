```python
# Exercise 1
from PyPDF2 import PdfReader

# Path to the PDF file without tables
file_path = "D:\\ALPHA\\Dynamic Folder\\Bellevue\\Winter 2023\\Data Mining\\Week 3\\Week_3_No_Tables.pdf"

# Reading the PDF file
try:
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    print("Text extracted from the PDF:")
    print(text)
except Exception as e:
    print(f"Error reading the PDF file: {e}")
```

```python
# Exercise 2 w/ PyPDF2 (as tabula did not work due to Java requirement)
import PyPDF2

# Path to the PDF file with tables
file_path = "D:\\ALPHA\\Dynamic Folder\\Bellevue\\Winter 2023\\Data Mining\\Week 3\\Week_3_With_Tables.pdf"

# Open the PDF file
with open(file_path, 'rb') as file:
    
# Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(file)
    
# Read through each page and extract text
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text = page.extract_text()
        if text:
            print(f"--- Page {page_num + 1} ---")
            print(text)
        else:
            print(f"--- Page {page_num + 1} has no text ---")
```

```python
# Exercise 4
# Import the relevant libraries

import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
from PIL import Image
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# File path to the image
image_path = r"D:\ALPHA\Dynamic Folder\Bellevue\Winter 2023\Data Mining\Week 3\bowers.jpg"

# Preprocess the image by converting to grayscale
image = Image.open(image_path).convert('L')

# Perform OCR using pytesseract
text = pytesseract.image_to_string(image)

# Replace newlines with spaces and remove hyphenated line breaks
clean_text = text.replace('\n', ' ').replace('- ', '')

# Print the cleaned text
print("Cleaned Text:")
print(clean_text)

# Process the cleaned text using spaCy
doc = nlp(clean_text)

# Print tokens with their POS tags and dependencies
print("\nTokens with POS and Dependency:")
for token in doc:
    print(f"{token.text}, {token.pos_}, {token.dep_}")
```
