##### Clean the “Tweet Content” column by removing non-text data and stop words.

```python
# Solution to Question 1

import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Load the CSV file
def load_csv(file_path):
    return pd.read_csv(file_path)

# Function to clean a single tweet
def clean_single_tweet(tweet):
    # Remove URLs
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    # Remove user @ references and '#' from tweet
    tweet = re.sub(r'\@\w+|\#', '', tweet)
    # Remove punctuations and numbers
    tweet = re.sub(r'[^A-Za-z]', ' ', tweet)
    # Remove single characters
    tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', tweet)
    return tweet

# Clean 'Tweet Content' column
def clean_tweet_content(data):
    # Set of English stopwords
    stop_words = set(stopwords.words('english'))

    cleaned_tweets = []
    for tweet in data:
        # Clean tweet
        clean_tweet = clean_single_tweet(tweet)
        # Tokenize the tweet
        word_tokens = word_tokenize(clean_tweet)
        # Remove stopwords
        filtered_tweet = [word for word in word_tokens if word not in stop_words]
        # Rejoin words
        cleaned_tweets.append(' '.join(filtered_tweet))
    return cleaned_tweets

# Main execution
def main():
    file_path = "D:\\ALPHA\\Dynamic Folder\\Bellevue\\Winter 2023\\Data Mining\\Week 5\\twitter_sample.csv"
    
    # Load data
    tweets_data = load_csv(file_path)

    # Clean 'Tweet Content'
    cleaned_tweets = clean_tweet_content(tweets_data["Tweet Content"])

    # Display the first few cleaned tweets
    for tweet in cleaned_tweets[:5]:
        print(tweet)

# Execute the main function
if __name__ == "__main__":
    main()
```

##### Filtering only tweets (not re-tweets) use your class from part one of this exercise to build BOW and TF-IDF Vectorizer representations of the text; print your results. Don't over-think this, leverage what the author does in the text.

```python
# Solution to Question 2

import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Load the CSV file
def load_csv(file_path):
    return pd.read_csv(file_path)

# Function to clean a single tweet
def clean_single_tweet(tweet):
    # Remove URLs
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    # Remove user @ references and '#' from tweet
    tweet = re.sub(r'\@\w+|\#', '', tweet)
    # Remove punctuations and numbers
    tweet = re.sub(r'[^A-Za-z]', ' ', tweet)
    # Remove single characters
    tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', tweet)
    return tweet

# Clean 'Tweet Content' column
def clean_tweet_content(data):
    # Set of English stopwords
    stop_words = set(stopwords.words('english'))

    cleaned_tweets = []
    for tweet in data:
        # Clean tweet
        clean_tweet = clean_single_tweet(tweet)
        # Tokenize the tweet
        word_tokens = word_tokenize(clean_tweet)
        # Remove stopwords
        filtered_tweet = [word for word in word_tokens if word not in stop_words]
        # Rejoin words into a clean tweet
        cleaned_tweets.append(' '.join(filtered_tweet))
    return cleaned_tweets

# Filter out retweets
def filter_retweets(data):
    return data[~data.str.startswith('RT')]

# Create a Bag-of-Words model
def create_bow(cleaned_tweets):
    vectorizer = CountVectorizer()
    bow_matrix = vectorizer.fit_transform(cleaned_tweets)
    print("Bag-of-Words Model:")
    print(bow_matrix.toarray())
    print(vectorizer.get_feature_names_out())

# Create a TF-IDF representation
def create_tfidf(cleaned_tweets):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(cleaned_tweets)
    print("TF-IDF Model:")
    print(tfidf_matrix.toarray())
    print(vectorizer.get_feature_names_out())

# Main execution function
def main():
    file_path = "D:\\ALPHA\\Dynamic Folder\\Bellevue\\Winter 2023\\Data Mining\\Week 5\\twitter_sample.csv"
    
    # Load data
    tweets_data = load_csv(file_path)

    # Clean 'Tweet Content'
    cleaned_tweets = clean_tweet_content(tweets_data["Tweet Content"])

    # Filter out retweets
    filtered_tweets = filter_retweets(pd.Series(cleaned_tweets))

    # Create and display Bag-of-Words model
    create_bow(filtered_tweets)

    # Create and display TF-IDF model
    create_tfidf(filtered_tweets)

# Execute the main function
if __name__ == "__main__":
    main()

```
